```python
# Function to ask user if they want to input different numbers
import math
def again():
    reply=input("Would you like to enter different numbers to calculate? [Y/N]  ").upper()
    if reply=="Y":
        run_calculator()
    elif reply=="N":
        print("Have a good day mate")
    else:
        print("Wrong answer")
   
```

```python
operation_dict={
    0: "Exit",
    1: "Addition",
    2: "Subtraction",
    3: "Multiplication",
    4: "Division",
    5: "Factorial",
    6: "Greatest Common Factor",
    7: "Least Common Multiple",
    8: "Area of Circle",
    9: "Cosine of Number",
    10: "Sine of Number",
    11: "Square Root of Number",
    12: "1st Number to Power of 2nd Number",
    13: "Degrees to Radians",
    14: "Hypotenuse of Triangle",
    15: "Natural Logarithm of Number"
}
 
       
```

```python
for keys,values in operation_dict.items():
    print (keys,": "+ values)
```

```python
def run_calculator():
    
    class calc():

        def __init__(self,a,b):
            self.a=a
            self.b=b

        def add(self):
            return self.a+self.b

        def subtract(self):
            return self.a-self.b

        def multiply(self):
            return self.a*self.b

        def divide(self):
            return self.a/self.b

        # greatest common factor
        def gcd(self):
            return math.gcd(self.a,self.b)

    
        def factorial(self):
            fact=1
            for i in range(1,self.a+1):
                fact=fact*i
            print(f"The factorial of {self.a} is {fact}")
            fact=1
            for i in range(1,self.b+1):
                fact=fact*i
            print(f"The factorial of {self.b} is {fact}")

        def lcm(self):
            return math.lcm(self.a,self.b)

        def area_of_circle(self):
            return self.a** 2 * 3.14159
    
        def cosine(self):
            return math.cos(self.a)

        def sine(self):
            return math.sin(self.a)

        def sqrt(self):
            return math.sqrt(self.a)

        def power(self):
            return math.pow(self.a,self.b)

        def deg_to_radians(self):
            return math.radians(self.a)

        def hypotenuse(self):
            c_squared= (self.a**2) + (self.b**2)
            return math.sqrt(c_squared)

        def ln(self):
            return math.log(self.a)
    
    a=int(input("Enter the first number: "))    
    b=int(input("Enter the second number (arbitrary if you only want to calculate the first number): "))

  # CREATING THE MY_CALC INSTANCE
    my_calc=calc(a,b)
# While loop
    choice=1
    while choice!=0:
        for keys,values in operation_dict.items():
            print (keys,": "+ values)
        print()     
# conditional statemetns
        choice=int(input("Enter the number of the calculation..."))
        print(f"Your numbers are {a} and {b}")
        print(f"You chose: {operation_dict[choice]}")
        print("------------------------")

        if choice==1:
            print(f"The sum is {my_calc.add()}")
            
        elif choice==2:
            print(f"The difference is {my_calc.subtract()}")
           
        elif choice==3:
            print(f"The product is {my_calc.multiply()}")
            
        elif choice==4:
            print(f"The quotient is {my_calc.divide()}")
            
        elif choice==5:
            my_calc.factorial()
        elif choice==6:
            print(f"The greatest common factor of {a} and {b} is {my_calc.gcd()}")
            
        elif choice==7:
            print(f"The least common multiple of {a} and {b} is {my_calc.lcm()}")
            
        elif choice==8:
            print(f"{a} is the radius of this circle. \nThe area of this circle is {my_calc.area_of_circle()} units squared")
            
        elif choice==9:
            print(f"The cosine of {a} is {my_calc.cosine(): .3f}")
            
        elif choice==10:
            print(f"The sine of {a} is {my_calc.sine(): 0.3f}")
            
        elif choice==11:
            print(f"The square root of {a} is {my_calc.sqrt(): 0.4f}")
            
        elif choice==12:
            print(f"{a} raised to the power of {b} is {my_calc.power()}")
            
        elif choice==13:
            print(f"{a} degrees translates to {my_calc.deg_to_radians(): 0.3f} radians ")
        
        elif choice==14:
            print(f"The hypotenuse of this triangle is {my_calc.hypotenuse()}")
            
        elif choice==15:
            print(f"The natural log of {a} is {my_calc.ln(): 0.3f}")
            
        elif choice==0:
            print("Exit")
        else:
            print("Sorry, invalid choice")
        print("------------------------")

    print() 
    
    again()
```

```python
run_calculator()
```
